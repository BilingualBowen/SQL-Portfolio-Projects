# Merge, Join, and Concatenate

```python
import pandas as pd
```

```python
LOTR_csv = input('Enter the path of the file LOTR.csv')
```

```python
df1 = pd.read_csv(LOTR_csv)
df1
```

```python
LOTR_2_csv = input('Enter the path of the file LOTR 2.csv')
```

```python
df2 = pd.read_csv(LOTR_2_csv)
df2
```

```python
# Merge

# Default - Inner join
df1.merge(df2)
```

```python
# Inner join on FellowshipID
df1.merge(df2, how = 'inner', on = 'FellowshipID')
```

```python
# Inner join on FellowshipID and FirstName
df1.merge(df2, how = 'inner', on = ['FellowshipID','FirstName'])
```

```python
# Outer join
df1.merge(df2, how = 'outer')
```

```python
# Left outer join
df1.merge(df2, how = 'left')
```

```python
# Right outer join
df1.merge(df2, how = 'right')
```

```python
# Cross join
df1.merge(df2, how = 'cross')
```

```python
# Join function

# Outer join
df1.join(df2, on='FellowshipID', how='outer', lsuffix='_Left', rsuffix='_Right')
```

```python
# Set FellowshipID as index while using join function

# Outer join
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix='_Left', rsuffix='_Right', how = 'outer')
df4
```

```python
# Concatenation

pd.concat([df1,df2])
```

```python
# Concat with inner join
pd.concat([df1,df2], join = 'inner')
```

```python
# Concat with outer join
pd.concat([df1,df2], join = 'outer')
```

```python
# Concat with outer join with axis=1, which joins by columns
pd.concat([df1,df2], join = 'outer', axis=1)
```

```python
# Append

df1.append(df2)
```
