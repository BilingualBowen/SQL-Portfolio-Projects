# Pandas Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(Ice_Cream_Ratings_csv)
```

```python
df
```

```python
df = df.set_index('Date')
```

```python
df
```

```python
# Use matplotlib.pyplot to plot in different styles
print(plt.style.available)
```

```python
# Use different matplotlib.pyplot styles to plot differently of dataframes below
plt.style.use('seaborn-v0_8-deep')
df.plot()
```

```python
# Use different matplotlib.pyplot styles to plot differently of dataframes below
plt.style.use('fivethirtyeight')
df.plot()
```

```python
# Use different matplotlib.pyplot styles to plot differently of dataframes below
plt.style.use('classic')
df.plot()
```

```python
# Use default plot style
plt.style.use('default')

# Pandas dataframe defult plot is (kind = 'line')

df.plot()
```

```python
# Pandas dataframe plot with subplots
df.plot(kind = 'line', subplots = True)
```

```python
# Pandas dataframe plot with title
df.plot(kind='line', title ='Ice Cream Ratings', xlabel = 'Daily Ratings', ylabel = 'Scores')
```

```python
# Pandas dataframe bar plot
df.plot(kind='bar')
```

```python
# Pandas dataframe stacked bar plot
df.plot(kind='bar', stacked = True)
```

```python
# Pandas dataframe single column plot
df['Flavor Rating'].plot(kind='bar')
```

```python
# Pandas dataframe plot horizontal bar chart
df.plot.barh(stacked=True)
```

```python
# Pandas dataframe scatter plot
df.plot.scatter(x='Texture Rating', y = 'Overall Rating', s = 200, c = 'Red')
```

```python
# Pandas dataframe histogram plot
# Histograms are good for showing distribution of variables
df.plot.hist(bins = 20)
```

```python
# Pandas dataframe boxplot
# Bottom-up: min, Q1(25th percdntile), median, Q3(75th percdntile), max
df.boxplot()
```

```python
# Pandas dataframe area plot
df.plot.area()
```

```python
# Plot with figsize
df.plot.area(figsize=(10,5))
```

```python
# Pandas dataframe pie plot
df.plot.pie(y='Flavor Rating', figsize=(10,10))
```
